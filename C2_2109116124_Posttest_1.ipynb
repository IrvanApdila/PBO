### Data Understanding

```python
# Dari dataset diatas buatlah  variabel yang berisi
# df1 isinya dataset dengan label target AttiredCustomer
# df2 dengan label bukan Attrired customer
# df3 dengan tabel untuk usia dibawah 50 tahun
# df4 untuk customer yang statusnya itu single

# membuat variabel df1 yang berisi dataset dengan label target AttiredCustomer
df1 = df.loc[df['Attrition_Flag'] == 'Attrited Customer']

# membuat variabel df2 yang berisi dataset dengan label bukan Attired Customer
df2 = df.loc[df['Attrition_Flag'] == 'Existing Customer']

# membuat variabel df3 yang berisi dataset untuk usia dibawah 50 tahun
df3 = df[df['Customer_Age'] < 50]

# membuat variabel df4 yang berisi dataset untuk customer yang statusnya single
df4 = df[df['Marital_Status'] == 'Single']

```

```python
# menggabungkan keempat dataframe menjadi satu dataframe
df = pd.concat([df1, df2, df3, df4], axis=0)

# menampilkan data df
print(df)
```

```python
# Tampilkan info dataset (df.info()) dan berikan penjelasannya dalam bentuk teks atau comment
print(df.info())
```

DataFrame tersebut terdiri dari 20 baris dan 10 kolom. Berikut adalah penjelasan detail beserta dengan tipe data masing-masing kolom pada DataFrame tersebut:

Unnamed: 0 : kolom integer yang berisi nomor urut atau indeks baris. Tipe datanya adalah int64.

CLIENTNUM : kolom integer yang berisi nomor identifikasi unik untuk setiap pelanggan bank. Tipe datanya adalah int64.

Attrition_Flag : kolom string yang berisi label target yang menunjukkan apakah pelanggan telah berhenti menjadi nasabah bank atau masih menjadi nasabah bank. Nilai yang mungkin adalah "Existing Customer" dan "Attrited 

Customer". Tipe datanya adalah object.

Customer_Age : kolom integer yang berisi usia pelanggan bank. Tipe datanya adalah int64.

Gender : kolom string yang berisi jenis kelamin pelanggan. Nilai yang mungkin adalah "M" untuk laki-laki dan "F" untuk perempuan. Tipe datanya adalah object.

Education_Level : kolom string yang berisi tingkat pendidikan pelanggan. Nilai yang mungkin adalah "High School", "Graduate", "Uneducated", "Unknown", "College", "Post-Graduate", dan "Doctorate". Tipe datanya adalah object.

Marital_Status : kolom string yang berisi status pernikahan pelanggan. Nilai yang mungkin adalah "Married", "Single", "Unknown", "Divorced", dan 

"Separated". Tipe datanya adalah object.

Income_Category : kolom string yang berisi kategori penghasilan pelanggan. Nilai yang mungkin adalah "Less than $40K", "$40K - $60K", "$60K - $80K", "$80K - $120K", "$120K +", dan "Unknown". Tipe datanya adalah object.

Card_Category : kolom string yang berisi jenis kartu kredit pelanggan. Nilai yang mungkin adalah "Blue", "Silver", "Gold", dan "Platinum". Tipe datanya adalah object.

Credit_Limit : kolom float yang berisi batas kredit pelanggan. Ini adalah jumlah maksimum uang yang dapat dipinjam oleh pelanggan menggunakan kartu kredit. Tipe datanya adalah float64.
Selain itu, informasi lain yang disajikan adalah jumlah nilai yang tidak kosong pada setiap kolom (non-null count) dan jumlah total baris (RangeIndex).
